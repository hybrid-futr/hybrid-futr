# Exploring the NL West with pybaseball

A brief analysis of the NL West division in 2022. Teams: Diamondbacks, Dodgers, Giants, Padres, Rockies

```python
import pybaseball as pyball
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Get all NL West Team Records for 2022
year = 2022
dbacks = pyball.schedule_and_record(year, 'ARI')
dodgers = pyball.schedule_and_record(year, 'LAD')
giants = pyball.schedule_and_record(year, 'SF')
padres = pyball.schedule_and_record(year, 'SD')
rockies = pyball.schedule_and_record(year, 'COL')
```
