## Exploring the NL with pybaseball

A brief analysis of the NL EAST, CENTRAL and WEST divisions in 2022. For the 2022 season, the playoff format for each lague has changed so there are now six teams from each lague in the postseason. The six teams for each league include each division champion, as well as three wild card selections which are the teams with the best record in the league that didn't outright win their division.

Each division ended the season with the following rankings:

- EAST: *Mets*>*Braves*>***Phillies***>Marlins>Nationals  
- CENTRAL: *Cardinals*>Brewers>Cubs>Reds>Pirates 
- WEST: *Dodgers*>*Padres*>Giants>Diamondbacks>Rockies

The teams are listed alphabetically here as follows: 

- EAST: Braves, Marlins, Mets, Nationals, Phillies  
- CENTRAL: Brewers, Cardinals, Cubs, Pirates, Reds 
- WEST: Diamondbacks, Dodgers, Giants, Padres, Rockies

UPDATE: The **Phillies** who finished the season *3rd in their division* and with the *worst record of all six* league postseason teams are going to the [World Series](https://www.mlb.com/postseason/history/world-series). Baseball is wild.

```python
import pybaseball as pyball
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Get all NL Team Records for 2022

year = 2022

# EAST
braves = pyball.schedule_and_record(year, 'ATL')
marlins = pyball.schedule_and_record(year, 'MIA')
mets = pyball.schedule_and_record(year, 'NYM')
nationals = pyball.schedule_and_record(year, 'WSN')
phillies = pyball.schedule_and_record(year, 'PHI')

#CENTRAL
brewers = pyball.schedule_and_record(year, 'MIL')
cardinals = pyball.schedule_and_record(year, 'STL')
cubs = pyball.schedule_and_record(year, 'CHC')
pirates = pyball.schedule_and_record(year, 'PIT')
reds = pyball.schedule_and_record(year, 'CIN')

#WEST
dbacks = pyball.schedule_and_record(year, 'ARI')
dodgers = pyball.schedule_and_record(year, 'LAD')
giants = pyball.schedule_and_record(year, 'SF')
padres = pyball.schedule_and_record(year, 'SD')
rockies = pyball.schedule_and_record(year, 'COL')
```

## Wins, Win Percentages, and Games Back

Comparing the wins, win percentages, and games back for each team of the NL West.

First we gather each team's Wins, Losses, and Win Percentages for the season.

```python
# Gets wins-to-date, losses-to-date, and win_percent-to-date for each team
dbacks['Wins'] = np.where(dbacks['W/L']=='W', 1, (np.where(dbacks['W/L']=='W-wo', 1, 0))).cumsum()
dbacks['Losses'] = np.where(dbacks['W/L']=='L', 1, (np.where(dbacks['W/L']=='L-wo', 1, 0))).cumsum()
dbacks['Win_Percentage'] = dbacks['Wins'] / (dbacks['Wins'] + dbacks['Losses'])

dodgers['Wins'] = np.where(dodgers['W/L']=='W', 1, (np.where(dodgers['W/L']=='W-wo', 1, 0))).cumsum()
dodgers['Losses'] = np.where(dodgers['W/L']=='L', 1, (np.where(dodgers['W/L']=='L-wo', 1, 0))).cumsum()
dodgers['Win_Percentage'] = dodgers['Wins'] / (dodgers['Wins'] + dodgers['Losses'])

giants['Wins'] = np.where(giants['W/L']=='W', 1, (np.where(giants['W/L']=='W-wo', 1, 0))).cumsum()
giants['Losses'] = np.where(giants['W/L']=='L', 1, (np.where(giants['W/L']=='L-wo', 1, 0))).cumsum()
giants['Win_Percentage'] = giants['Wins'] / (giants['Wins'] + giants['Losses'])

padres['Wins'] = np.where(padres['W/L']=='W', 1, (np.where(padres['W/L']=='W-wo', 1, 0))).cumsum()
padres['Losses'] = np.where(padres['W/L']=='L', 1, (np.where(padres['W/L']=='L-wo', 1, 0))).cumsum()
padres['Win_Percentage'] = padres['Wins'] / (padres['Wins'] + padres['Losses'])

rockies['Wins'] = np.where(rockies['W/L']=='W', 1, (np.where(rockies['W/L']=='W-wo', 1, 0))).cumsum()
rockies['Losses'] = np.where(rockies['W/L']=='L', 1, (np.where(rockies['W/L']=='L-wo', 1, 0))).cumsum()
rockies['Win_Percentage'] = rockies['Wins'] / (rockies['Wins'] + rockies['Losses'])
```

## Graphing Wins and Win Percentages

```python
# Graph Wins Comparison
plt.rcParams["figure.figsize"] = (8,6)

plt.plot(dbacks['Wins'], label='Diamondbacks', c='red')
plt.plot(dodgers['Wins'], label='Dodgers', c='blue')
plt.plot(giants['Wins'], label='Giants', c='orange')
plt.plot(padres['Wins'], label='Padres', c='gold')
plt.plot(rockies['Wins'], label='Rockies', c='purple')

plt.xticks(np.arange(0, len(dodgers.index), step=10))
plt.xlabel('Game')
plt.ylabel('Wins')

plt.legend(loc='lower right')

plt.title("NL West Wins Comparison ({})".format(year))
```

```python
# Graph Win Percentage Comparison #
plt.rcParams["figure.figsize"] = (8,6)

plt.plot(dbacks['Win_Percentage'], label='Diamondbacks', c='red')
plt.plot(dodgers['Win_Percentage'], label='Dodgers', c='blue')
plt.plot(giants['Win_Percentage'], label='Giants', c='orange')
plt.plot(padres['Win_Percentage'], label='Padres', c='gold')
plt.plot(rockies['Win_Percentage'], label='Rockies', c='purple')

plt.xticks(np.arange(0, len(dodgers.index), step=10))
plt.xlabel('Game')
plt.ylabel('Win Percentage')

plt.legend(loc='lower right')

plt.title("NL West Win Percentage Comparison ({})".format(year))
```
